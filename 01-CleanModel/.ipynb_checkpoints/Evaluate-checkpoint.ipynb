```python
import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader
from torchvision import models, transforms
from torchvision.models import VGG16_BN_Weights
from torchvision import transforms
from PIL import Image
import pandas as pd
from pathlib import Path
from sklearn.metrics import classification_report, confusion_matrix, f1_score
from tqdm import tqdm
import numpy as np
import matplotlib.pyplot as plt 

class CSVDataset(Dataset):
    def __init__(self, csv_file, img_dir, class_to_idx, transform=None):
        self.df = pd.read_csv(csv_file, low_memory=False)
        self.img_dir = Path(img_dir)
        self.class_to_idx = class_to_idx
        self.transform = transform

    def __len__(self):
        return len(self.df)

    def __getitem__(self, idx):
        row = self.df.iloc[idx]
        rel = row['rel_path']
        img_path = self.img_dir / rel.lstrip('/')
        image = Image.open(img_path).convert('RGB')
        if self.transform:
            image = self.transform(image)
        label = self.class_to_idx[row['class']]
        return image, label, rel

def _load_checkpoint_weights(model, checkpoint_path, map_location='cpu', strict=True):
    ckpt = torch.load(checkpoint_path, map_location=map_location)
    if isinstance(ckpt, dict):
        if 'model_state_dict' in ckpt:
            state = ckpt['model_state_dict']
        elif 'state_dict' in ckpt:
            state = ckpt['state_dict']
        else:
            state = ckpt
    else:
        state = ckpt

    new_state = {}
    for k, v in state.items():
        new_key = k
        if k.startswith('module.'):
            new_key = k[len('module.'):]
        new_state[new_key] = v

    try:
        model.load_state_dict(new_state, strict=strict)
    except Exception as e:
        print(f"Warning: strict load failed ({e}). Trying non-strict load (missing/extra keys will be ignored).")
        model.load_state_dict(new_state, strict=False)

def evaluate_model(
    model,
    checkpoint_path,
    detect_csv,
    image_dir,
    class_to_idx,
    device=None,
    batch_size=32,
    img_size=128,
    mean=(0.4812775254249573, 0.4674863815307617, 0.4093940854072571),
    std=(0.19709135591983795, 0.1933959424495697, 0.19051066040992737),
    out_csv_path=None,
    verbose=True
):

    if device is None:
        device = torch.device('cuda:0' if torch.cuda.is_available() else 'cpu')
    else:
        device = torch.device(device)

    detect_transform = transforms.Compose([
        transforms.ToTensor(),
        transforms.Normalize(mean, std)
    ])

    dataset = CSVDataset(detect_csv, image_dir, class_to_idx, transform=detect_transform)
    dataloader = DataLoader(dataset, batch_size=batch_size, shuffle=False,
                            pin_memory=(device.type=='cuda'))

    model = model.to(device)

    _load_checkpoint_weights(model, checkpoint_path, map_location=device)

    model.eval()
    all_preds = []
    all_labels = []
    all_rels = []

    with torch.no_grad():
        for inputs, labels, rels in tqdm(dataloader, desc="Evaluating", leave=verbose):
            inputs = inputs.to(device)
            outputs = model(inputs)
            if outputs.dim() == 1:
                preds = (outputs > 0.5).long()
            else:
                _, preds = torch.max(outputs, dim=1)
            all_preds.extend(preds.cpu().numpy().tolist())
            all_labels.extend(labels.numpy().tolist())
            all_rels.extend(rels)

    idx_to_class = {v: k for k, v in class_to_idx.items()}
    classes_sorted = [idx_to_class[i] for i in sorted(idx_to_class.keys())]

    report = classification_report(all_labels, all_preds, target_names=classes_sorted, zero_division=0)
    cm = confusion_matrix(all_labels, all_preds)
    f1 = f1_score(all_labels, all_preds, average='macro', zero_division=0)

    if verbose:
        print("\n=== Classification Report ===")
        print(report)
        print("=== Confusion Matrix ===")
        print(cm)
        print(f"\nMacro F1: {f1:.4f}")

    plt.figure(figsize=(8, 6))
    plt.imshow(cm, cmap=plt.cm.Blues)
    plt.title('Confusion Matrix')
    plt.colorbar()
    plt.xticks(range(len(classes_sorted)), classes_sorted, rotation=45, ha='right')
    plt.yticks(range(len(classes_sorted)), classes_sorted)
    thresh = cm.max() / 2.0 if cm.size else 0.0
    for i, j in np.ndindex(cm.shape):
        plt.text(j, i, int(cm[i, j]), ha='center', va='center',
                 color='white' if cm[i, j] > thresh else 'black')
    plt.ylabel('True')
    plt.xlabel('Pred')
    plt.tight_layout()
    plt.show()

    out_df = pd.DataFrame({
        'rel_path': all_rels,
        'true_idx': all_labels,
        'pred_idx': all_preds,
    })
    out_df['true_class'] = out_df['true_idx'].map(idx_to_class)
    out_df['pred_class'] = out_df['pred_idx'].map(idx_to_class)

    if out_csv_path:
        out_csv_path = Path(out_csv_path)
        out_csv_path.parent.mkdir(parents=True, exist_ok=True)
        out_df.to_csv(out_csv_path, index=False)
        if verbose:
            print(f"Saved detailed results to {out_csv_path}")

    return {
        'report_text': report,
        'confusion_matrix': cm,
        'macro_f1': f1,
        'per_image_df': out_df
    }
```

```python
DETECT_CSV = './Dataset/detection_model.csv'

class_to_idx = {c: i for i, c in enumerate(['bird', 'cat', 'fish', 'frog', 'monkey', 'moth', 'spider'])}

print (class_to_idx)

def image_dir(size):
    return Path(f'./Dataset/{size}x{size}/')
```

```python
model = models.vgg16_bn(weights=VGG16_BN_Weights.IMAGENET1K_V1)
model.avgpool = nn.AdaptiveAvgPool2d((7,7))
model.classifier[6] = nn.Linear(4096, 7)
```

```python
IMG_SIZE = 32

metrics = evaluate_model(
    model=model,
    checkpoint_path='./Models/32x32/best_min_acc_vgg16_32x32_Model-2.pth',
    detect_csv=DETECT_CSV,
    image_dir=image_dir(IMG_SIZE),
    class_to_idx=class_to_idx,
    batch_size=128,        
    img_size=IMG_SIZE,        
    out_csv_path=Path(f"./Evaluate/detection_results_{IMG_SIZE}x{IMG_SIZE}.csv"),
    verbose=True
)
```

# Compare

```python
import pandas as pd
import matplotlib.pyplot as plt

sizes = [32, 64, 128, 256, 512, 1024]
csv_files = [f"./Evaluate/detection_results_{x}x{x}.csv" for x in sizes]
dfs = [pd.read_csv(f) for f in csv_files]

correct_sets = []
for df in dfs:
    correct_paths = set(df.loc[df['true_class'] == df['pred_class'], 'rel_path'])
    correct_sets.append(correct_paths)

correct_in_all_paths = set.intersection(*correct_sets)

df_all_correct = dfs[0][dfs[0]['rel_path'].isin(correct_in_all_paths)]

class_counts = df_all_correct['true_class'].value_counts()

print(f"Number of correct predictions in all models: {len(df_all_correct)}")


df_all_correct.to_csv("./Evaluate/correct_in_all_models.csv", index=False)

plt.figure(figsize=(8,5))
class_counts.plot(kind='bar')
plt.xlabel("Class")
plt.xticks(rotation=45)  
plt.ylabel("Number of Images Correct in All Models")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.savefig("./Evaluate/histogram_correct_all_models.png")
plt.show()

df_all_correct
```
